```python
import numpy as np
import statsmodels.tsa.stattools as stattools
import statsmodels.tsa.statespace.tools as tools
import matplotlib.pyplot as plt

import screener as scr
```

```python
mu = .1
sig = .3
x = np.random.normal(mu,sig, 99)
y = np.zeros(100)
for i in range(1,100):
    y[i] = y[i-1]+x[i-1]
```

```python
plt.plot(y)
```

```python
y_diff = tools.diff(y)
```

```python
plt.plot(y_diff)
```

```python
np.mean(y_diff)
```

```python
np.std(y_diff)
```

```python
result = stattools.adfuller(y_diff)
```

```python
result[1]
```

```python
import numpy as np
import networkx as nx
import statsmodels.tsa.stattools as stattools
import statsmodels.tsa.statespace.tools as tools
import matplotlib.pyplot as plt
import market.screener as screener
import market.iex as iex
import scipy.stats as stats
```

Let's get some stock price series data for the last 300 trading days.

```python
screen = screener.Screener(['AMZN', 'AMD', 'FB', 'GOOGL', 'MSFT', 'AAPL', 'IETC', 'HYG', 'IGSB'], 300)
```

Now we will look at the 'trend_dict' method. Since the market has been volatile lately, I will use AMD from May to September of 2018 as an example. We will save the original AMD series for later.

```python
oldAMD = screen.series_dict['AMD']
newAMD = screen.series_dict['AMD'][0:100]
screen.series_dict['AMD'] = newAMD
```

```python
trend_dict = screen.trend_dict()
```

```python
trend_dict
```

So AMD has a pretty big drift at .01 
Now lets use the whole AMD series and use the coint_graph method to find cointegrated stocks

```python
screen.series_dict['AMD'] = oldAMD
```

```python
G = screen.coint_graph(.1)
```

```python
pos=nx.get_node_attributes(G,'pos')
labels = nx.get_edge_attributes(G,'weight')
nx.draw(G,pos)
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
```

So now we could look at the connected components of our graph

```python
cc = list(nx.connected_components(G))
cc
```

```python
tiks = [screen.encoding_dict[i] for i in cc[4]]
tiks
```

```python
for tik in tiks:
    plt.figure(figsize=(10,5))
    plt.title(tik+str())
    plt.plot(screen.series_dict[tik])
```

These two tickers' price history indeed share similar features

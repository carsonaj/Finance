```python
connection = iex.Connection()
s = connection.get_series('IETC', 100)
s = np.log(s)
x = s.values
t = s.index
diff = tools.diff(s)
```

```python
plt.figure(num=1, figsize=(20,10))
plt.subplot(211)
plt.plot(t,x)
plt.subplot(212)
diff.plot()
```

```python
print(np.mean(diff))
print(np.std(diff))
stattools.adfuller(diff)
```

```python
x_train = x[:80]
x_test = x[80:]
```

```python
model = arima_model.ARIMA(endog=x_train, order=(2,1,1))
results = model.fit()
results.summary()
```

```python
print(results.conf_int(alpha=.1))
```

```python
forecast = results.forecast(steps=20, alpha=.2)
```

```python
plt.figure(num=2, figsize=(20,10))
plt.subplot(211)
plt.plot(t[80:], x_test)
plt.plot(t[80:], forecast[0], color='red')
plt.plot(t[80:], forecast[2][:,0], color='orange')
plt.plot(t[80:], forecast[2][:,1], color='orange')
plt.subplot(212)
plt.plot(t, x)
plt.plot(t[80:], forecast[0], color='red')
plt.plot(t[80:], forecast[2][:,0], color='orange')
plt.plot(t[80:], forecast[2][:,1], color='orange')
```
